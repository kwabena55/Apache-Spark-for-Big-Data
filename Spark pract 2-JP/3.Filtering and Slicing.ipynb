# Import Libraries and Read Data 

```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
spark=SparkSession.builder.appName("Filtering and Slicing 2").getOrCreate()
```

```python
df=spark.read.csv('/Users/solansah/Desktop/Data/appl_stock.csv',header=True,inferSchema=True)
df.show()
```

```python
df.printSchema()
```

```python
df.show(5)
```

```python
df.select("Open").show()
```

```python
df.filter("close < 500").show()
```

```python
df.filter("Open > 200")
```

```python
df.filter("Open > 200").select(["Open","Close","Volume"]).show()
```

```python
df.filter(df['Close'] < 500).show()
```

# Filtering based on Multiple Conditions

```python
df.filter((df["Close"] > 200) & (df['Close'] < 300)).show()
```

```python
df.filter("Low ==197.16").show()
```

```python
df.filter("Low==197.16").select("Date").show()
```

# Collect 
* Collect shows data as tuples in a list 

```python
df.filter("Low==197.16").select("Date").collect()
```

```python
df.filter("Low==197.16").collect()
```

```python
for i in df.filter("Low==197.16").collect():
    print(i[-1])
```

```python
result=df.filter("Low==197.16")
result.show()
```

```python
df.filter("Low > 100").collect()[0].asDict()
```

```python
df.filter("Low > 100").collect()[0:3]
```
