# .Collect Method ()
* As you recall .show() prints a dataframe for us but for .collect it stores them in a form of a list

```python
df.filter(df["Low"]== 209.020004).collect()
```

```python
df.filter(b  &  ~ d).collect()
```

```python
df.filter(df["Low"]== 209.020004).collect()[0]
```

```python
df.filter(df["Low"]== 197.16).collect()  # In a lsit
```

```python
df.filter(df["Low"]== 197.16).collect()[0] # not in a list
```

# Change to Dict

```python
rows=df.filter(df["Low"]== 197.16).collect()[0] # not in a list
rows.asDict()
```
